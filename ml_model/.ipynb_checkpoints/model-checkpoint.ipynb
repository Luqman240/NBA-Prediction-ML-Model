```python
import os
import pandas as pd
import numpy as np
```

```python
DATAFRAMES_PATH = '../DATAS/'
SEASONS = range(2018,2024)
```

```python
#Load all excel and join in one dataframe
def load_dataframes():
    full_dataframe = pd.DataFrame(dtype=float)
    for season in SEASONS:
        df = pd.read_excel(DATAFRAMES_PATH+f'season_{season}.xlsx')
        df["SEASON"] = season
        full_dataframe = pd.concat([full_dataframe, df], ignore_index=True)
        
    print("Datas imported successfully !")
    return full_dataframe
```

```python
DATAFRAME = load_dataframes()
```

```python
#Charlotte Bobcats got renamed Charlotte Hornets in 2014-2015 so i have to change CHO to CHA
DATAFRAME['TEAM'].value_counts()
```

```python
DATAFRAME['TEAM'] = DATAFRAME['TEAM'].replace("CHO", "CHA")
len(DATAFRAME['TEAM'].value_counts())
```

```python
del DATAFRAME["MP"]
del DATAFRAME["OPP_MP"]
```

```python
DATAFRAME
```

```python
#Add the target value 
def add_target(team):
    team["TARGET"] = team["WON"].shift(-1)
    return team

df = DATAFRAME.groupby("TEAM", group_keys=False).apply(add_target)
```

```python
df[df["TEAM"] == "MEM"]
```

```python
df[df["TEAM"] == "MEM"]
```

```python
df["TARGET"][pd.isnull(df["TARGET"])] = 2
```

```python
df["TARGET"] = df["TARGET"].astype(int, errors="ignore")
```

```python
df[df["TEAM"] == "MEM"] 
```

```python
#Scaling datas
unscalable_column = ["WON", "TARGET", "TEAM", "TEAM_OPP", "SEASON"]
scalable_column = df.columns[~df.columns.isin(unscalable_column)]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_scores = scaler.fit_transform([df["PTS"]])
df[scalable_column] = scaler.fit_transform(df[scalable_column]) #Scaling datas between 0 and 1 to improve ML Model Performance
```

```python
""" MACHINE LEARNING """
```

```python
from sklearn.model_selection import TimeSeriesSplit #CANT USE BASIC CROSS VALIDATION SPLIT BC CASE WHERE THE FUTURES OBS ARE TRAINED TO PREDICT THE PAST
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=5)
sfs = SequentialFeatureSelector(rr, n_features_to_select=20 ,direction="backward", cv=split)
```

```python
sfs.fit(df[scalable_column], df["TARGET"]) #Find what features make prediction better
```

```python
predictors = list(scalable_column[sfs.get_support()])
```

```python
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []
    seasons = sorted(df["SEASON"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        
        train = data[data["SEASON"] < season] #Get All Datas before current season
        test = data[data["SEASON"] == season]
        
        model.fit(train[predictors], train["TARGET"])
        preds = model.predict(test[predictors])
        
        preds = pd.Series(preds, index=test.index) #Change np.array to df for next step
        
        combined = pd.concat([test["TARGET"], preds], axis=1) #Df who shows actual value and the predicted one
        combined.columns = ["ACTUAL", "PREDICTION"]
        
        all_predictions.append(combined)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, rr, predictors)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score

predictions = predictions[predictions["ACTUAL"] != 2]
accuracy_score(predictions["ACTUAL"], predictions["PREDICTION"])
```

```python
def find_team_averages(team): #Get the mean performance of a team for the last 10 games
    numeric_columns = team.select_dtypes(include='number').columns
    rolling = team
    rolling[numeric_columns] = team[numeric_columns].rolling(10).mean()

    return rolling

df_rolling = df_rolling.groupby(["TEAM", "SEASON"], group_keys=False).apply(find_team_averages)
```
